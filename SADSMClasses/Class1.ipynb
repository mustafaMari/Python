```python
import matplotlib.pyplot as plt
x = [1,6,7,8,5,4,5,3,65,345,76,5,34,67,5,34,654,3,4,65,34,35]

print(len(x))
y = [2,5,1,8,4,6,9,64,2,34,23,2,435,23,56,3,54,34,54,54,34,57]

print(len(y))
r = []
for xi,yi in zip(x,y):
    r.append(float((xi*yi) / xi * xi))
plt.plot(r)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.show()
```
